# Conjunto de Prueba WFG

```python
import matplotlib.pyplot as plt
import numpy as np
import glob  # Para leer mltiples archivos

# Ruta de los archivos de texto
ruta_archivos = "data/*.dat" 

# Obtener lista de archivos
archivos = sorted(glob.glob(ruta_archivos))

# Configurar la figura
plt.figure(figsize=(8, 6))

# Leer y graficar cada archivo
for archivo in archivos:
    data = np.loadtxt(archivo)  # Carga el archivo como un array de NumPy
    plt.figure(figsize=(6, 5))
    plt.plot(data[:, 0], data[:, 1], label=archivo.split("/")[-1], marker='.', linewidth=0.8, linestyle='')  # Puntos x, y

    # Configuracin del grfico
    plt.xlabel("Objetivo 1")
    plt.ylabel("Objetivo 2")
    plt.title(f"Frente de Pareto - {archivo.split('/')[-1]}")
    plt.legend()  # Muestra nombres de archivos en la leyenda
    plt.grid(True)

    # Mostrar grfico
    plt.show()
```
